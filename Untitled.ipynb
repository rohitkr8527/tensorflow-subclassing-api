## <b> Data Preparation

```python
import tensorflow as tf
from tensorflow.keras import layers, Model
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Load and normalize CIFAR-10
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
y_train = y_train.flatten()
y_test = y_test.flatten()

# Class names for display
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']
```

## <b> Subclassed Model 

```python
class CIFAR10Model(Model):
    def __init__(self):
        super(CIFAR10Model, self).__init__()
        self.conv1 = layers.Conv2D(32, 3, activation='relu')
        self.pool1 = layers.MaxPooling2D(2)
        self.conv2 = layers.Conv2D(64, 3, activation='relu')
        self.pool2 = layers.MaxPooling2D(2)
        self.flatten = layers.Flatten()
        self.dense1 = layers.Dense(128, activation='relu')
        self.dropout = layers.Dropout(0.5)
        self.out = layers.Dense(10)

    def call(self, x, training=False):
        x = self.conv1(x)
        x = self.pool1(x)
        x = self.conv2(x)
        x = self.pool2(x)
        x = self.flatten(x)
        x = self.dense1(x)
        if training:
            x = self.dropout(x, training=training)
        return self.out(x)

model = CIFAR10Model()
```

## <b> Custom Training Loop

```python
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
optimizer = tf.keras.optimizers.Adam()
train_acc = tf.keras.metrics.SparseCategoricalAccuracy()
test_acc = tf.keras.metrics.SparseCategoricalAccuracy()

train_ds = tf.data.Dataset.from_tensor_slices((x_train, y_train)).shuffle(50000).batch(64)
test_ds = tf.data.Dataset.from_tensor_slices((x_test, y_test)).batch(64)

EPOCHS = 10

for epoch in range(EPOCHS):
    print(f"\nEpoch {epoch+1}/{EPOCHS}")
    
    # Training
    for images, labels in train_ds:
        with tf.GradientTape() as tape:
            logits = model(images, training=True)
            loss = loss_fn(labels, logits)
        grads = tape.gradient(loss, model.trainable_variables)
        optimizer.apply_gradients(zip(grads, model.trainable_variables))
        train_acc.update_state(labels, logits)

    print(f"Train Accuracy: {train_acc.result().numpy():.4f}")
    train_acc.reset_state()

    # Testing
    for test_images, test_labels in test_ds:
        logits = model(test_images, training=False)
        test_acc.update_state(test_labels, logits)

    print(f"Test Accuracy: {test_acc.result().numpy():.4f}")
    test_acc.reset_state()
```

## <b> Save and Load Model 

```python
# Save model
model.save('cifar10_model.keras') 
# Alternatively, you could use .h5 extension for HDF5 format: model.save('cifar10_model.h5')

# To reload later:
# model = tf.keras.models.load_model('cifar10_model.keras')
```

## <b> Normalized Confusion Matrix

```python
# Gather predictions and true labels
y_pred, y_true = [], []

for images, labels in test_ds:
    logits = model(images, training=False)
    preds = tf.argmax(logits, axis=1)
    y_pred.extend(preds.numpy())
    y_true.extend(labels.numpy())

y_pred = np.array(y_pred)
y_true = np.array(y_true)

# Compute confusion matrix and normalize it
cm = confusion_matrix(y_true, y_pred, normalize='true')
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)

plt.figure(figsize=(10, 8))
disp.plot(cmap=plt.cm.Blues, xticks_rotation=45)
plt.title("Normalized Confusion Matrix on CIFAR-10")
plt.show()
```

## <b> Predictions

```python
plt.figure(figsize=(10, 10))
for i in range(9):
    idx = np.random.randint(len(x_test))
    img = x_test[idx]
    true_label = y_test[idx]

    logits = model(np.expand_dims(img, axis=0), training=False)
    pred_label = tf.argmax(logits, axis=1).numpy()[0]

    plt.subplot(3, 3, i + 1)
    plt.imshow(img)
    plt.title(f"True: {class_names[true_label]}\nPred: {class_names[pred_label]}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
